```python
place_list = ['B-03','B-37','C-16','C-17','C-18','C-19','C-20','C-21','D-07','D-05','F-05']
```

```python
import pandas as pd
df = pd.DataFrame()

for i in range(1,13):
    df = df.append(pd.read_excel(str(i)+'월 서울시 교통량 조사자료(2022).xlsx', sheet_name = 1))
```

```python
new_df = df.loc[df['지점번호'].isin(place_list)]
```

```python
new_df = new_df.sort_values(by = ["지점번호","방향","일자"])
new_df
```

```python
new_df.loc[new_df['지점번호'] == 'D-05','방향'] = new_df[new_df['지점번호'] == 'D-05']['방향'].replace({'유입':'유출','유출':'유입'})
```

```python
new_df['지점번호_방향'] = new_df['지점번호'].map(str) + "" + new_df['방향'].map(str)
```

```python
new_df = new_df.drop(['요일','지점명','지점번호','방향','구분'], axis = 1)
```

```python
new_df = pd.melt(new_df, id_vars=['지점번호_방향','일자'])
```

```python
new_df.loc[new_df["variable"] == "0시","variable"] = "00시"
new_df['time'] = new_df['일자'].map(str) + "" + new_df.variable.str.rstrip("시")
```

```python
new_df = new_df.drop(['일자','variable'], axis = 1)
```

```python
new_df.time = pd.to_datetime(new_df.time, format ='%Y%m%d%H')
```

```python
new_df.sort_values(by = ['지점번호_방향','time'])
```

```python
new_df = new_df.pivot(index='time', columns='지점번호_방향', values='value')
```

```python
new_df.to_csv('통행량_2022.csv', encoding = "cp949")
```
